```python
top_nationalities = df['Nationality'].value_counts().head(5)

plt.figure(figsize=(8,5))
top_nationalities.plot(kind='bar', color='yellow', edgecolor='black')

plt.title('Top 5 Nationalities in FIFA Dataset')
plt.xlabel('Nationality')
plt.ylabel('Number of Players')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
position_map = {
    'CB': 'Defender', 'LB': 'Defender', 'RB': 'Defender', 'LWB': 'Defender', 'RWB': 'Defender', 'LCB': 'Defender', 'RCB': 'Defender',
    'CDM': 'Midfielder', 'CM': 'Midfielder', 'RCM': 'Midfielder', 'LCM': 'Midfielder', 'CAM': 'Attacker', 'LM': 'Winger', 'RM': 'Winger',
    'LDM': 'Defender', 'RDM': 'Defender', 'LAM': 'Attacker', 'RAM': 'Attacker', 'LW': 'Winger', 'RW': 'Winger', 'ST': 'Striker', 'CF': 'Attacker',
    'RF': 'Attacker', 'LF': 'Attacker', 'RS': 'Striker', 'LS': 'Striker'
}

df['Role'] = df['Position'].map(position_map)


df = df.dropna(subset=['Role'])


sns.violinplot(data=df, x='Role', y='Overall', palette='muted')
plt.title("Player Overall Distribution by Position")
plt.ylabel("Overall")
plt.xlabel("Player Position")
plt.show()
```

```python
position_to_role = {
    'CB': 'Defender', 'LB': 'Defender', 'RB': 'Defender', 'LWB': 'Defender', 'RWB': 'Defender', 'LCB': 'Defender', 'RCB': 'Defender',
    'CDM': 'Midfielder', 'CM': 'Midfielder', 'RCM': 'Midfielder', 'LCM': 'Midfielder', 'CAM': 'Attacker', 'LM': 'Winger', 'RM': 'Winger',
    'LDM': 'Defender', 'RDM': 'Defender', 'LAM': 'Attacker', 'RAM': 'Attacker', 'LW': 'Winger', 'RW': 'Winger', 'ST': 'Striker', 'CF': 'Attacker',
    'RF': 'Attacker', 'LF': 'Attacker', 'RS': 'Striker', 'LS': 'Striker'
}

df['Role'] = df['Position'].map(position_to_role)
import seaborn as sns
import matplotlib.pyplot as plt

role_counts = df['Role'].value_counts().reset_index()
role_counts.columns = ['Role', 'Player Count']

plt.figure(figsize=(8, 5))
sns.barplot(data=role_counts, x='Role', y='Player Count', palette='viridis')
plt.title('Number of Players by Broader Roles')
plt.xlabel('Role')
plt.ylabel('Player Count')
plt.tight_layout()
plt.show()
```
